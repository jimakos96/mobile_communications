```python
#Κόσυβας Δημήτριος  ΑΜ:03114828
#Θ1
import numpy as np 
import matplotlib.pyplot as plt
from scipy.stats import norm
from pynverse import inversefunc
#α)
Lmax = 128 #dB
hms = 1.28 #m
Pthr = -128 #dBm

hbs=[] 
for i in range (30,201): #για το COST 231 - Hata μοντέλο για το hbs πρέπει να ισχύει 30m <= hbs <= 200m
    hbs.append(i) # σε m

fc = 1850 # σε MHz
a_u = 3.2*(np.log10(hms*11.75))**2 - 4.97 #dB
a_r = hms*(1.1*np.log10(fc)-0.7) - 1.56*np.log10(fc) + 0.8 #dB
Au = 46.3 + 33.9*np.log10(fc) - 13.82*np.log10(hbs) - a_u
B = (44.9-6.55*np.log10(hbs))
Ar = 46.3 + 33.9*np.log10(fc) - 13.82*np.log10(hbs) - a_r
C_u = 3 #dB
C_r = 0 # dB
d_u=10**((Lmax-Au-C_u)/B)
d_r=10**((Lmax-Ar-C_r)/B)
print("Για αστικές περιοχές:")
for i in range (30,201):
        print ("Για ύψος σταθμού βάσης",hbs[i-30],"m, η μέγιστη απόσταση κάλυψης είναι",d_u[i-30],"km")
```

```python
print("Για αγροτικές περιοχές:")
for i in range (30,201):
        print ("Για ύψος σταθμού βάσης",hbs[i-30],"m, η μέγιστη απόσταση κάλυψης είναι",d_r[i-30],"km")
```

```python
plt.figure(figsize=(20,10))
plt.plot(hbs,d_u,'b',label="urban")
plt.plot(hbs,d_r,'r',label="rural")
plt.xlim(30,200)
plt.title("hbs-dmax for urban and rural areas")
plt.xlabel("hbs(m)")
plt.ylabel("dmax(km)")
plt.legend(loc="upper left")
plt.grid()
plt.show()
```

```python
#β)
# Με βάση τη σύσταση ΙΤU-R P.1406-1:
dh = 8 #interdecile height variation (m) 
lamda = 300/fc #m
#L=L50+Ls όπου L50 προκύπτει από το Cost 231- Hata μοντέλο και το Ls απώλειες λόγω shadowing
sl = 2.7 + 0.42*np.log10(fc/100) + 1.01*(np.log10(fc/100))**2 #dB-standard deviation of the location variability distribution
#Επιλέγω αυτόν τον τύπο σε κάθε περίπτωση διότι έχουμε μέγιστη αποστάση < 50 km και εν τέλει δεν εισάγω το ύψος στον τύπο.
Au2 = 46.3 + 33.9*np.log10(fc) - 13.82*np.log10(30) - a_u #Για hbs = 30 m (η μικρότερη αποδεκτή τιμή για το Cost 231 - Hata)
Au3 = 46.3 + 33.9*np.log10(fc) - 13.82*np.log10(200) - a_u #Για hbs = 200 m (η μεγαλύτερη αποδεκτή τιμή για το Cost 231 - Hata)
Ar2 = 46.3 + 33.9*np.log10(fc) - 13.82*np.log10(30) - a_r
Ar3 = 46.3 + 33.9*np.log10(fc) - 13.82*np.log10(200) - a_r
B2 = (44.9-6.55*np.log10(30))
B3 = (44.9-6.55*np.log10(200))
Pav=[] #θέτω τιμές από 0.9 έως 0.999 για την πιθανότητα κάλυψης 
for i in range (9000,10000):
    i=i/10000
    Pav.append(i)
ο=np.ones(1000)
d_u_30=10**((Lmax-Au2-C_u-sl*inversefunc(norm.sf,y_values=ο-Pav))/B2) #sl*inversefunc(norm.sf,y_values=ο-Pav) : τύπος που 
#συνδέει το Ls με το Pav
d_u_200=10**((Lmax-Au3-C_u-sl*inversefunc(norm.sf,y_values=ο-Pav))/B3)
d_r_30=10**((Lmax-Ar2-C_r-sl*inversefunc(norm.sf,y_values=ο-Pav))/B2)
d_r_200=10**((Lmax-Ar3-C_r-sl*inversefunc(norm.sf,y_values=ο-Pav))/B3)
plt.figure(figsize=(20,10))
plt.plot(Pav,d_u_30,'b',label="urban")
plt.plot(Pav,d_r_30,'r',label="rural")
plt.xlim(0.9,1)
plt.legend(loc="upper right")

plt.title("Pav-dmax for urban and rural areas for hbs=30m")
plt.xlabel("Pav")
plt.ylabel("dmax(km)")
plt.grid()
plt.show()
plt.figure(figsize=(20,10))
plt.plot(Pav,d_u_200,'b',label="urban")
plt.plot(Pav,d_r_200,'r',label="rural")
plt.xlim(0.9,1)
plt.title("Pav-dmax for urban and rural areas for hbs=200m")
plt.xlabel("Pav")
plt.ylabel("dmax(km)")
plt.grid()
plt.legend(loc="upper right")
plt.show()
```

```python
#γ)

#Θα χρησιμοποιήσουμε την εξίσωση του Friis: Pr=Pt*Gt*Gr*(λ/4πd)**2 όπου Pr=Pthr προκειμένου να σιγουρέψουμε ότι υπάρχει επικοινωνία
#σε κάθε περίπτωση. Σ΄ύμφωνα με το προηγούμενο ερώτημα για Pav[500]=0.95,hbs=30m σε urban area το dmax είναι d_u_30[500].Παρακάτω
#δημιουργώ τα δεδομένα που χρειάζομαι με βάση αυτήν τη σκέψη και αντίστοιχα δουλεύω για κάθε περίπτωση.

lamda = 300/fc # m
Gt = 10**1.5
Gr = 10**0.2
Pthrlin = (10**-12.5)*(10**-3)

#95%
#Urban area:
du = []
for i in range(1,int(1000*d_u_30[500])):
    du.append(i)
Pu = np.zeros(len(du))
Pudbm = np.zeros(len(du))
Pu[0] = Pthrlin/(Gt*Gr*(lamda/(4*np.pi*1000*d_u_30[500]))**2) #W
Pudbm[0] = 10*np.log10(Pu[0]/(10**-3)) #dBm
for i in range (1,len(Pu)):
    Pu[i]=Pu[0]/(du[i]**2) #W - θεωρώ ότι η ισχύς πέφτει με ρυθμό 1/(d**2) αφού οι άλλες απώλειες έχουν ληφθεί 
                           #υπόψη στον υπολογισμό του dmax
    Pudbm[i] = 10*np.log10(Pu[i]/(10**-3))   #dBm                   
#Rural area:                                      
dr = []
for i in range (1,int(1000*d_r_30[500])):
    dr.append(i)
Pr = np.zeros(len(dr))
Prdbm = np.zeros(len(dr))
Pr[0] = Pthrlin/(Gt*Gr*(lamda/(4*np.pi*1000*d_r_30[500]))**2) #W
Prdbm[0] = 10*np.log10(Pr[0]/(10**-3))
for i in range (1,len(Pr)):
    Pr[i]=Pr[0]/(dr[i]**2)
    Prdbm[i] = 10*np.log10(Pr[i]/(10**-3))
du1= []
for i in range(1,int(1000*d_u_200[500])):
    du1.append(i)
Pu1 = np.zeros(len(du1))
Pudbm1= np.zeros(len(du1))
Pu1[0] =Pthrlin/(Gt*Gr*(lamda/(4*np.pi*1000*d_u_200[500]))**2)
Pudbm1[0] = 10*np.log10(Pu1[0]/(10**-3))
for i in range (1,len(Pu1)):
    Pu1[i]=Pu1[0]/(du1[i]**2)
    Pudbm1[i] = 10*np.log10(Pu1[i]/(10**-3))
dr1 = []
for i in range (1,int(1000*d_r_200[500])):
    dr1.append(i)
Pr1 = np.zeros(len(dr1))
Prdbm1= np.zeros(len(dr1))
Pr1[0] =Pthrlin/(Gt*Gr*(lamda/(4*np.pi*1000*d_r_200[500]))**2)
Prdbm1[0] = 10*np.log10(Pr1[0]/(10**-3))
for i in range (1,len(Pr1)):
    Pr1[i]=Pr1[0]/(dr1[i]**2)
    Prdbm1[i] = 10*np.log10(Pr1[i]/(10**-3))

    
    
plt.figure(figsize=(20,10))
plt.plot(du,Pudbm,'r',label="urban-hbs=30m")
plt.plot(dr,Prdbm,'g',label="urban-hbs=30m")
plt.plot(du1,Pudbm1,'y',label="rural-hbs=200m")
plt.plot(dr1,Prdbm1,'m',label="urban-hbs=200m")
plt.title("d-Pt for urban and rural areas,Pav=95%")
plt.xlabel("d(m)")
plt.ylabel("Pt(dBm)")
plt.legend(loc="upper right")
plt.grid()
plt.show()
```
